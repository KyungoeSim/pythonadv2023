```python
import requests #웹페이지 읽어오는 모듈
from bs4 import BeautifulSoup as bs #읽어오느 웹페이지 정제
import pandas as pd #데이터분석 모듈 필수!
```

```python
url = "https://naver.com/"
txt = requests.get(url) #읽어온 페이지가 의미가 없는 그냥 글자일 뿐
#print(txt)
```

```python
url = "https://naver.com/"
txt = requests.get(url)
html = bs(txt.text) #html로 변환
#print(html)
```

```python
#######################################
#벅스차트 읽어오기
#######################################
url = "https://music.bugs.co.kr/chart"
txt = requests.get(url)
html = bs(txt.text) #html로 변환
#print(html)
```

```python
len(html.select('tr'))

```

```python
len(html.select('tbody > tr'))
```

```python
len(html.select('table > tbody > tr'))
```

```python
len(html.select('table.byChart > tbody> tr')) #.점은 클래스를 의미한다.
```

```python
songs = html.select('table.byChart > tbody> tr')
print(songs)
```

```python
song=songs[0]
print(song)
```

```python
len(song.select('p>a'))
```

```python
len(song.select('p.title>a'))
```

```python
print(song.select('p.title>a'))
```

```python
#태그 그냥 버리고 글자만 뽑는법

print(song.select('p.title > a')[0].text)
```

```python
#2등 제목 뽑기
print(songs[1].select('p.title > a')[0].text)

#2등 가수명 뽑기
print(songs[1].select('p.artist > a')[0].text)
```

```python
#전체 100곡의 노래제목과 가수명 추출

song_data = []
rank = 1

for song in songs:
    
    title = song.select('p.title > a')[0].text
    singer = song.select('p.artist > a')[0].text
    song_data.append(['Bugs',rank,title,singer])
    rank += 1



df = pd.DataFrame(song_data, columns = ['서비스', '순위', '타이틀', '가수'])
df
```

```python
#df 엑셀로 변환
df.to_excel('bugs.xlsx', index=False)
```

```python
###########################################################################
#                       Melon 차트 읽어오기                                #
###########################################################################

#######################################
#멜론차트 읽어오기
#######################################
#url = "https://www.melon.com/chart/index.htm"
#txt = requests.get(url)
#html = bs(txt.text) #html로 변환
#print(html)

#크롤링 막고 있음
#다른방식으로 크롤링 시도해야함

from selenium import webdriver #위에서의 request 와는 다른 방식

driver = webdriver.Chrome('chromedriver.exe')
driver.get("https://www.melon.com/chart/index.htm")

txt = driver.page_source
html = bs(txt)

print(html)
```

```python
len(html.select('tr'))
```

```python
len(html.select('tbody > tr'))
```

```python
songs = html.select('tbody > tr')
```

```python
song = songs[0]

print(song)
```

```python
len(song.select('div.rank01>span>a'))
```

```python
print(song.select('div.rank01>span>a')[0].text)
```

```python
song_data = []
rank = 1

for song in songs:
    
    title = song.select('div.rank01>span>a')[0].text
    singer = song.select('div.rank02>span>a')[0].text
    song_data.append(['멜론',rank,title,singer])
    rank += 1



df = pd.DataFrame(song_data, columns = ['서비스', '순위', '타이틀', '가수'])
df
```

```python
df.to_excel('멜론.xlsx', index=False)
```
